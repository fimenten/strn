```python
import matplotlib.pyplot as plt
import numpy as np
import sklearn.linear_model
```

```python
x = np.linspace(-1,1,100)
y = np.random.standard_normal(x.shape)/10 + x ** 2
```

```python
reg_all = sklearn.linear_model.LinearRegression()
reg_all.fit(x.reshape((-1,1)),y)
res = (y - reg_all.predict(x.reshape(-1,1)))
print(np.mean(res**2))
xx = np.linspace(-1,1,1000)
yy = reg_all.predict(xx.reshape(-1,1))
plt.scatter(x,y)
plt.plot(xx,yy)
```

```python
reg_1 = sklearn.linear_model.LinearRegression()
reg_2 = sklearn.linear_model.LinearRegression()

x_1 = x[x>0]
y_1 = y[x>0]
reg_1.fit(x_1.reshape(-1,1),y_1)

reg_2 = sklearn.linear_model.LinearRegression()
x_2 = x[x<=0]
y_2 = y[x<=0]
reg_2.fit(x_2.reshape(-1,1),y_2)

res = (y - (reg_1.predict(x.reshape(-1,1))+ reg_2.predict(x.reshape(-1,1)))*0.5)
print(np.mean(res**2))
res = (y_1 - reg_1.predict(x_1.reshape(-1,1)))**2 + (y_2 - reg_2.predict(x_2.reshape(-1,1)))**2
print(res.mean())
```
